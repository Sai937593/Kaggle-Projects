```python
cat_cols = []
for col in train_df.columns:
  n= train_df[col].nunique()
  if n <= 10:
    cat_cols.append(col)
cat_cols
```

```python
train_df_encoded = pd.get_dummies(train_df, columns=cat_cols)
test_df_encoded = pd.get_dummies(test_df, columns=cat_cols)
test_df_encoded = test_df_encoded.drop(columns=['Polestar', 'smart'])
```

```python
set(train_df_encoded.columns).symmetric_difference(set(test_df_encoded.columns))
```

```python
for col in train_df_encoded.columns:
  if col not in cat_cols:
    print(col)
```

```python
brand_value_counts_dict = train_df_encoded.brand.value_counts().to_dict()
```

```python
train_df_encoded['brand_value_counts'] = train_df_encoded.brand.map(lambda x: np.log1p(brand_value_counts_dict[x]))
test_df_encoded['brand_value_counts'] = test_df_encoded.brand.map(lambda x: np.log1p(brand_value_counts_dict[x]))
```

```python
model_value_counts_dict_train = train_df_encoded.model.value_counts().to_dict()
model_value_counts_dict_test = test_df_encoded.model.value_counts().to_dict()
train_df_encoded['model_value_counts'] = train_df_encoded.model.map(lambda x: np.log1p(model_value_counts_dict_train[x]))
test_df_encoded['model_value_counts'] = test_df_encoded.model.map(lambda x: np.log1p(model_value_counts_dict_test[x]))
```

```python
X = train_df_encoded.drop(columns=['price'])
y = train_df_encoded.price
X.shape, y.shape
```

```python
X.isnull().any(), y.isnull().any()
```

```python
corr= abs(X.corrwith(y)).sort_values(ascending=False)
corr
```

```python
from sklearn.model_selection import cross_validate, train_test_split
from sklearn.metrics import mean_squared_error, make_scorer
from sklearn.linear_model import LinearRegression, Ridge, Lasso, SGDRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.svm import SVR

linreg = LinearRegression()
ridge = Ridge(alpha=200)
lasso = Lasso(alpha=200)
rf = RandomForestRegressor()
sgd = SGDRegressor( penalty='elasticnet', alpha=10, learning_rate='adaptive')
svr = SVR()

useful_cols = corr[corr>0.1].index

X_useful = X[useful_cols]
scaler2 = StandardScaler()
y_scaled = scaler2.fit_transform(y.values.reshape(-1, 1))
y_scaled = y_scaled.ravel()
X_train, X_test, y_train, y_test = train_test_split(X_useful, y_scaled, test_size=0.2, random_state=42)
print(f'using these columns only {useful_cols}')
print(f'X_train shape: {X_train.shape}')
print(f'y_train shape: {y_train.shape}')
print(f'y_train mean std: {np.mean(y_train), np.std(y_train), np.min(y_train), np.max(y_train)}')
models = (linreg, ridge, lasso, sgd, svr)
scores_dict = dict()
for model in models:
  print(f'fitting the model: {model}')
  pipeline = Pipeline([
    # ('poly', PolynomialFeatures(degree=3)),
    ('scaler', StandardScaler()),
    ('model', model)
  ])
  scores = cross_validate(estimator=pipeline, X=X_train, y=y_train, cv=2, scoring=make_scorer(mean_squared_error))
  scores_dict[model] = scores['test_score']

scores_dict
```
